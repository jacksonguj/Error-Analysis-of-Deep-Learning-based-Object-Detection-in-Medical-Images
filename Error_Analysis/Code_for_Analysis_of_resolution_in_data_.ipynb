# Check resolution distribution for Polyps data

```python
import os

Image_Path = '/content/drive/MyDrive/BSc_Project/Kvasir-SEG/images/'
images_nam = os.listdir(Image_Path)
```

```python
import PIL
from PIL import Image

resolutions = {}
for imge in images_nam:
  img = PIL.Image.open(Image_Path+imge)
  resolutions[imge] = img.size
  h,w = img.size
```

```python
import matplotlib.pyplot as plt

split_resolution = list(zip(*(list(resolutions.values()))))
plt.scatter(split_resolution[0],split_resolution[1],s = 10, marker = '*')
plt.xlabel('hight')
plt.ylabel('width')
plt.vlines (510, 350, 1100, colors='gray', linestyle='solid')
plt.hlines (455, 350, 1900, colors='gray', linestyle='solid')
plt.vlines (1150, 350, 1100, colors='gray', linestyle='solid')
plt.hlines (990, 350, 1900, colors='gray', linestyle='solid')
min_hight = min(split_resolution[0])
min_width = min(split_resolution[1])
```

```python
print(list(zip(resolutions.keys(),resolutions.values())))
```

```python
lower_resol = {}
higher_resol = {}
list_resolutions = list(zip(resolutions.keys(),resolutions.values()))

for img, resol_info in list_resolutions:
  h = resol_info[0]
  w = resol_info[1]
  if 510 > h and 455 > w:
    lower_resol[img] = resol_info
    print(resol_info)
  if w > 990 and h > 990:
    higher_resol[img] = resol_info
```

```python
print(len(lower_resol))
print(len(higher_resol))
```

# Check Resolution distribution in natural dataset (coco dataset)

```python
import os
import PIL
from PIL import Image

Image_Path = '/content/drive/MyDrive/BSc_Project/Natural_image/val2017/'
images_nam = os.listdir(Image_Path)

resolutions = []
lower_images = []
for imge in images_nam:
  img = PIL.Image.open(Image_Path+imge)
  resolutions.append(img.size)
  h,w = img.size
  if not w >= 512 and h >= 512:
    lower_images.append(imge)
```

```python
import matplotlib.pyplot as plt

split_resolution = list(zip(*resolutions))
min_hight = min(split_resolution[0])
min_width = min(split_resolution[1])

number_over_standard = 0

for h,w in resolutions:
  if h >= 512 and w >= 512:
    number_over_standard += 1

plt.scatter(split_resolution[0],split_resolution[1],s = 10, marker = '*')
plt.xlabel('hight')
plt.ylabel('width')
# plt.vlines (512, 350, 1100, colors='gray', linestyle='solid')
# plt.hlines (512, 350, 1900, colors='gray', linestyle='solid')

print('min_hight:',min_hight)
print('min_width:',min_width)
print('number_over_standard:',number_over_standard)
print('total_nuber_image:',len(resolutions))
```
